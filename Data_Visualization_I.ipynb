## Importando os Pacotes em Nosso Codigo:

```python
#importando o pacote
import seaborn as sns
import matplotlib.pyplot as plt

#vamos verificar os datasets de exemplo
help(sns.load_dataset)
```

## Selecionando um DataFrame de Exemplo:

```python
#vamos escolher o dataset de exemplo, chamado tips este nos mostra o valor da gorgeta dado de acordo varias caracteristicas de entrada:
data = sns.load_dataset('tips')
data.head()
```

## Visualizando os Dados:

Podemos visualizar dos dados, de diferentes formas com os pacotes SeaBorn e Matplotlib, no exemplo abaixo estamos analisando a dispersao entre as 
variáveis tip e total_bill referenciando os dias da semana:

```python
#visualizando os dados
sns.relplot(data=data
            , x="total_bill"
            , y="tip"
            , hue="day"
            , style="day" )
```

```python
#tambem podemos visualizar de forma separada e obter outros insights:
sns.relplot(data=data
           , x="total_bill"
           , y="tip"
           , hue="day"
           , style="day"
           , col="time"
           , row="sex" )
```

# Elementos de Um Gráfico:

Listamos Abaixo Os Principais Elementos de um Gráfico:

 - Tamanho da Figura.  
 - Títulos e Eixos.  
 - Legendas.  
 - Paleta de Cores.  



## Alterando o Tamanho dos Graficos:

```python
#vamos inicialmente criar um objeto da classe sns:
graph = sns.relplot(data=data
                    , x="total_bill"
                    , y="tip"
                    , hue="day"
                    , style="day" )

#conversão: 1 polegada = 2.54 cm
altura = 20 / 2.54
largura = 30 / 2.54

#Esta linha altera o tamanho do grafico
graph.fig.set_size_inches(w=largura, h=altura)
```

## Alterando Títulos e Eixos:

```python
#Titulos:
graph.ax.set_title("Gorjetas"                      #alterando o titulo.
                  , fontsize=12                    #alterando as fontes.
                  , fontweight="bold");            #alterando 'sombra' da fonte.

#Eixos:
graph.set_xlabels("Valor da conta (USD)"           #alterando o nome do eixo.
                 , fontsize=10);                   #alterando o tamanho da fonte.
                
graph.set_ylabels("Valor da gorjeta (USD)"         #alterando o nome do eixo.
                 , fontsize=10);                   #aletrando o tamanho fonte

#exibindo o grafico:
graph.fig
```

## Alterando a Legenda:

```python
graph = sns.relplot(data=data, x="total_bill", y="tip", hue="day", style="day")

graph.ax.set_title("Gorjetas", fontsize=12, fontweight="bold");
graph.set_xlabels("Valor da conta (USD)", fontsize=10);
graph.set_ylabels("Valor da gorjeta (USD)", fontsize=10);

graph.legend.set_title("Dias da semana", prop={"size": 10, "weight": "bold"});

#conversão: 1 polegada = 2.54 cm
altura = 20 / 2.54
largura = 30 / 2.54

#Esta linha altera o tamanho do grafico
graph.fig.set_size_inches(w=largura, h=altura)
```

## Alterando a paleta de cores:

```python
#declarando o grafico:
graph = sns.relplot(data=data, x="total_bill", y="tip", hue="day", style="day", palette = "rocket")

#set de legendas e eixos
graph.ax.set_title("Gorjetas", fontsize=12, fontweight="bold");
graph.set_xlabels("Valor da conta (USD)", fontsize=10);
graph.set_ylabels("Valor da gorjeta (USD)", fontsize=10);

#set de titulo
graph.legend.set_title("Dias da semana", prop={"size": 10, "weight": "bold"});

#conversão: 1 polegada = 2.54 cm
altura = 20 / 2.54
largura = 30 / 2.54

#Esta linha altera o tamanho do grafico
graph.fig.set_size_inches(w=largura, h=altura)
```
