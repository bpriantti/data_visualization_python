### Grafico de Barras:

Um gráfico de barras é nada mais nada menos que um diagrama usado na ora de comparar valores diferentes, as barras maiores representam valores mais altos. Os gráficos de barras também são conhecidos como gráficos de colunas e podem ser horizontais ou verticais.

Obs:
Eles são ideais para demonstrar a diferença de valor ou tamanho entre vários itens.

```python
#importando a biblio seaborn:
import seaborn as sns

#load do dataset tips:
data = sns.load_dataset("tips")

#verificando o dataset:
data.head()
```

```python
#agora vamos agregar os dados de total_bill por day, usando o metodo agregatte:
tips = data[["day", "total_bill"]].groupby("day").agg("sum").reset_index()

#verificando o dataset
tips.head()
```

```python
#vamos visualizar graficamente os dados:
graph = sns.barplot(data=tips, x="day", y="total_bill", ci=None, palette="pastel")

#definindo elementos do grafico:
graph.set(title='Valor da conta por dia da semana', xlabel='Dia da semana', ylabel='Valor da conta (USD)');
```

```python
#vamos visualizar na horizontal:

#declarando o grafico:
grafico = sns.barplot(data=tips, y="day", x="total_bill", ci=None, palette="pastel")

#defiindo elementos do grafico:
grafico.set(title='Valor da conta por dia da semana', ylabel='Dia da semana', xlabel='Valor da conta (USD)');
```

Comparando duas features em apenas um gráfico:

```python
#Exemplo: Valor da conta por dia da semana por período:

#agrupando por day e time:
tips = data[["day", "total_bill", "time"]].groupby(["day", "time"]).agg("sum").reset_index()

#visualizando o dataset:
tips.head(10)
```

```python
#declarando o grafico:
grafico = sns.barplot(data=tips                            #dados
                      , x="day"                            #elemento x
                      , y="total_bill"                     #elemento y
                      , hue="time"                         #coluna categorica
                      , palette="pastel")                  #paleta de cores

#definindo elementos do grafico:
grafico.set( title='Valor da conta por dia da semana por período'
            ,xlabel='Dia da semana'
            ,ylabel='Valor da conta (USD)');
```

### Gráfico de linha:

Um gráfico de linha é um tipo de gráfico utilizado para mostrar dados numéricos/quantitativos e este podem mostrar a relação de uma variável com o tempo, ou mais de uma variável.

```python
#importando a biblioteca seaborn
import seaborn as sns

#load do dataset
data = sns.load_dataset("flights")

#inspecionando o dataset
data.head()
```

```python
#agregando por ano e numero de passageiros:

#declarando o dataset agregao
flights = data[["year", "passengers"]].groupby("year").agg("sum").reset_index()

#inspecionando
flights.head()
```

```python
#declarando o line plot:
graph = sns.lineplot(data=flights, x="year", y="passengers", palette="pastel")

#definindo elementos do grafico:
graph.set(title='Passageiros por ano', xlabel='Ano', ylabel='Passageiros');
```

```python
#declarando o grafico:
graph = sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="pastel")

#implementando elementos do grafico:
graph.set(title='Passageiros por mês por ano', xlabel='Mês', ylabel='Passageiros');

#implementando a legenda para fora do grafico:
graph.get_legend().set_title("Ano");
sns.move_legend(grafico, "upper left", bbox_to_anchor=(1, 1));
```

```python
#tambem podemos realizar query's que basicamente sao requisicoes para selecionar anos com base em um criterio:
flights = data.query("1955 <= year < 1960")

#declarando o grafico:
graph = sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="pastel")

#definindo os elementos do grafico:
graph.set(title='Passageiros por mês por ano', xlabel='Mês', ylabel='Passageiros');

#implementando a legenda para fora do grafico:
graph.get_legend().set_title("Ano");
sns.move_legend(graph, "upper left", bbox_to_anchor=(1, 1));
```

### Gráfico de Area:

O gráfico de área e um gráfico semelhante ao gráfico de linha, mas acrescenta uma hachura embaixo da linha mostrada.

```python
#importando a biblioteca
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
data = sns.load_dataset("flights")

#inspecionando dataset
data.head()
```

```python
#agregando anos e passageiros:
flights = data[["year", "passengers"]].groupby("year").agg("sum").reset_index()

#inspecionando o dataset:
flights.head()
```

```python
#declarando o grafico:
grafico = sns.FacetGrid(data=flights, palette="pastel")

#configurando area do grafico:
grafico.map(sns.lineplot, "year", "passengers")
grafico.map(plt.fill_between, 'year', 'passengers', alpha=0.3)

#definindo elementos do grafico:
grafico.set(title='Passageiros por ano', xlabel='Ano', ylabel='Passageiros');

#configurando o tamanho do grafico:
grafico.fig.set_size_inches(w=15/2.54, h=7.5/2.54)
```

### Histograma:

O histograma representa a distribuição de uma variável numérica. A variável numérica é segmentada em intervalos representados por uma barra. Já a altura da barra indica a contagem dos valores presentes na segmentação.

```python
#importando a biblitoeca seaborn
import seaborn as sns
```

```python
#realizando o load do dataset
data = sns.load_dataset("penguins")

#verificando o dataset
data.head()
```

```python
#declarando o grafico de histograma:
graph = sns.histplot(data=data, x="flipper_length_mm",bins = 20)

#declarando os elementos com o tipo:
graph.set(title='Dist. flipper_length_mm', xlabel='Valor flipper_length_mm', ylabel='Contagem');
```

### Kernel Density Estimator - (KDE):

O kernel Density Estimator se preocupa em aproximar os dados do histograma de uma distribuição, sendo também uma boa forma de visualizar os dados.


```python
#declarando a biblioteca
import seaborn as sns

#load dataset
data = sns.load_dataset("titanic")

#visualizando o dataset
data.head()
```

```python
#avisualizando a distribuicao das colunas age e survived:
titanic = data[["age", "survived"]]

#declarando o dataset
grafico = sns.histplot(data=titanic, x="age", hue="survived", kde=True)

#definindo elementos do grafico
grafico.set(title='Idade por sobrevivência', xlabel='Idade', ylabel='Contagem');

#defindo titulo da legenda
grafico.get_legend().set_title("Sobreviveu?");
```

```python
#tambem podemos visualizar tres distribuicoes diferentes:

#selecionando as colunas:
titanic = data[["age", "survived", "class"]]

#declarando o grafico:
grafico = sns.FacetGrid(data=titanic, col="class",  hue="survived")

#selecionando o kde
grafico.map(sns.histplot, "age", kde=True);
```

## Gráfico de Dispersão:

O gráfico de dispersão representa a dispersão entre duas variáveis numéricas. Cada valor é representado um ponto P(x,y). É útil para observar a variação conjunta de duas variáveis.

```python
#importando a biblioteca seaborn
import seaborn as sns

#load dataset
data = sns.load_dataset("tips")

#visualizando o grafico de dispersao de total_bill por tip, que represetam o valor total da conta e a gorgeta
sns.scatterplot(data=data, x="total_bill", y="tip");
```

```python
#tambem podemos visualizar duas dispersoes juntas duspersas por categoria os dados:
sns.scatterplot(data=data, x="total_bill", y="tip", hue="time");

#aqui conseguimos visualizar a difenca dos dados por almoco(lunch) e dinner(janta)
```

## Mapa de Calor:

O mapa de calor representação a correlação entre três variáveis, essencialmente uma representação 2D de um gráfico 3D. Cada valor é representado por um ponto com três coordenadas: x e y indicam a sua posição e o z (necessariamente numérica) a sua intensidade. É útil para observar a distribuição geral dos dados.

```python
import seaborn as sns

data = sns.load_dataset("flights")
data.head()

flights = data.pivot("month", "year", "passengers")
flights.head()

grafico = sns.heatmap(data=flights, cmap="Spectral")
grafico.set(title='Passageiros por mês por ano', xlabel='Ano', ylabel='Mês');
```

```python
grafico = sns.heatmap(data=flights, cmap="Spectral", annot=True, fmt="d")
grafico.set(title='Passageiros por mês por ano', xlabel='Ano', ylabel='Mês');
grafico.figure.set_size_inches(w=20/2.54, h=10/2.54)
```

```python
data = sns.load_dataset("flights")

flights = data.pivot("year", "month", "passengers")
flights.head()

grafico = sns.heatmap(data=flights, cmap="Spectral")
grafico.set(title='Passageiros por mês por ano', xlabel='Mês', ylabel='Ano');
```

# Gráficos em 3 eixos:


Podemos também visualizar gráficos em 3 dimensões, utilizando a biblioteca matplotlib, mais a frente vou abordar tópicos mais avançados sobre este assunto no estando vou deixar um exemplo, mostrando um plot em três dimensoes de superfície matemática.

caso queiram se aprofundar:

https://www.askpython.com/python-modules/matplotlib/3-dimensional-plots-in-python#:~:text=How%20to%20Plot%203-Dimensional%20Plots%20in%20Python%3F%201,wire-frame.%20...%207%20Create%20a%20surface%20plot.%20

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
#create 3d axes
fig = plt.figure()
ax = plt.axes(projection='3d')
 
#function for Z values
def f(x, y): 
    return np.cos(np.sqrt(x ** 2 + y ** 2)) 
   
# x and y values 
x = np.linspace(1, 10, 10) 
y = np.linspace(1, 10, 10) 
   
X, Y = np.meshgrid(x, y) 
Z = f(X, Y) 
   
ax = plt.axes(projection ='3d') 
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')

plt.show()
```
